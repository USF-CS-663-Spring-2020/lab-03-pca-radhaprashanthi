```python
feature_df = df.pattern.str.split(' ', expand=True)
```

```python
#Changing object type to float and int
numfeature_df = feature_df.apply(pd.to_numeric)
```

```python
numfeature_df.info()
```

```python
numfeature_df
```

```python
#Replacing NaN with -1 to avoid discrepencies
pattern_df = numfeature_df.fillna(-1)
```

```python
pattern_df
```

```python
#Normalizing the data using StandardScaler
from sklearn.preprocessing import StandardScaler

x = pattern_df.values
x = StandardScaler().fit_transform(x)
x
```

```python
import numpy as np

np.mean(x), np.std(x)
```

The data has a normal distribution with mean = 0 and std = 1

```python
from sklearn.decomposition import PCA

#considering only 2 principal components
pca_gisette = PCA(n_components=2)
principalComponents_gisette = pca_gisette.fit_transform(x)
principal_gisette_df = pd.DataFrame(data = principalComponents_gisette, columns = ['PC1','PC2'] )
principal_gisette_df
```

```python
pca_gisette.explained_variance_ratio_
```

Total variance ratio explained by the two principle components is 5.14%.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xlabel('PC 1')
plt.ylabel('PC 2')
targets = [1, -1]
colors = ['r','b']
for target,color in zip(targets,colors):
    indicesToKeep = label_df == target
    #print(indicesToKeep)
    plt.scatter(principal_gisette_df.loc[indicesToKeep['diagnosis'], 'PC1'],
                principal_gisette_df.loc[indicesToKeep['diagnosis'], 'PC2'],
                c = color)
plt.legend(targets,prop={'size': 15})
```

How many principal components do we need to explain the variance in data?

```python
# Variance graph with first 50 components
pca = PCA().fit(x[:50])
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Total Explained Variance');
```

As we can see in the above graphs, as the components are increasing the contribution to variance is increasing but after 2800 components, it remains constant.
